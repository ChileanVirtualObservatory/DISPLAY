# Detection and classification of spectroscopic lines

The algorithm has as purpose the identification of chemical components of the astronomical object to study. For this, we analyze its spectroscopic lines to find patterns and predict its composition.
For an astronomical object, observe their spectroscopic lines may help identify the molecules that compose it, since for each state of this molecule will emit energy peaks at certain frequencies.
If there is confusion between two molecules for a line being rated, it is possible to give a probabilistic prediction of molecule which corresponds to the line for the presence of each molecule along the entire measured spectrum.

### Sparse Coding

$$
min_{\alpha} ||x-D\alpha||_2^2 
s.a. 
||\alpha||_1 \leq \lambda
$$

$x$ : observed vector, 

$D$ : recalibrated dictionary, 

$\lambda$ : spartity parameter, 


$||()||_2$ : l-2 norm

```python
y_train = get_fortran_array(np.asmatrix(X))
dictionary_recal_fa = np.asfortranarray(dictionary_recal, dtype= np.double)

param = {
  'lambda1' : 25, # practically unrestricted = 1000
  # 'lambda2' : 0.1,
  # 'L': 1,
  'pos' : True,
  'mode' : 0,
  'ols' : True,
  'numThreads' : -1} # number of cores to use; the default choice is -1
# and uses all the cores of the machine
alpha = spams.lasso(y_train, dictionary_recal_fa, **param).toarray()
total = np.inner(dictionary_recal_fa, alpha.T)
```
